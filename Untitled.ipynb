```python
class Human:
    def __init__(self, age, name, mood = "good"):
        self.age = age
        self.name = name
        self.mood = mood
    def getInfo(self):
        print(f'My name is {str(self.name)}, I am {self.age} years old!')

mick = Human(26, "Mick", "sehr gut")
mick.getInfo()

class Mutant(Human):
    def __init__(self, age, name, superpower, mood = "gut"):
        super().__init__(name=name, age=age, mood=mood)
        self.superpower = superpower
    def __str__(self):
        return f'Ich bin ein Mutant und mein Name ist {self.name}'
    def getSuperpower(self):
        print(f'Mein Name ist {str(self.name)} und ich kann {self.superpower}!')
mick_super = Mutant(age=26, name = "Mick", superpower = "Fliegen", mood = "sehr gut")
mick_super.getSuperpower()
mick_super.getInfo()
print(mick_super)
```
